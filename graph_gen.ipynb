```python
import math
import matplotlib.pyplot as plt
sides = [2, 3, 4, 5, 6, 7]

phash = []
shash = []
p8rep = []
s8rep = []

for side in sides:
    cells = side * side
    primary_cells = math.ceil(cells / 2)
    secondary_cells = math.floor(cells / 2)
    ph = (6 ** primary_cells)
    sh = (6 ** secondary_cells)
    pr = (8 ** primary_cells)
    sr = (8 ** secondary_cells)
    phash.append(ph)
    shash.append(sh)
    p8rep.append(pr)
    s8rep.append(sr)


# Plot how faster representation grows than combinations (use log scale)
plt.yscale("log")
# Name the axes
plt.xlabel("Side length")
plt.ylabel("Number of combinations of representation")
# Plot the data
plt.plot(sides, phash, label="Primary hash")
plt.plot(sides, p8rep, label="Primary 3bit per cell")
plt.legend()    


# And another plot for secondary hash
plt.figure()
plt.yscale("log")
plt.xlabel("Side length")
plt.ylabel("Number of combinations of representation")
plt.plot(sides, shash, label="Secondary hash")
plt.plot(sides, s8rep, label="Secondary 3bit per cell")
plt.legend()
```
